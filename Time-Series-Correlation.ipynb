```python
import pandas as pd 
import numpy as np

from sklearn import preprocessing

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('stocks/total_data_price_Change_Symbol.csv', index_col = 'Symbol')
df.head()
```

```python
def calculate_timeseries_correlation(data):
    result = []
    days = data.shape[1]
    len_symbols = data.shape[0]
    for i in range(len_symbols):
        symbol_corr = []
        for j in range(len_symbols):
            
            if i == j :
                symbol_corr.append(int(days / 2))
                continue
                
                
            change = [1 if np.sign(data[i , day]) == np.sign(data[j , day])
                      else 0 
                      for day in range(days)]

            symbol_corr.append(sum(change))

        result.append(symbol_corr)
    return result
```

```python
symbols = df.index
data = df.to_numpy()
correlation = calculate_timeseries_correlation(data)
```

```python
df_corr = pd.DataFrame(correlation, columns = symbols , index = symbols) 
df_corr.to_csv('symbols-correlation.csv')
df_corr.head()
```

```python
x = df_corr.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))
x_scaled = min_max_scaler.fit_transform(x)
df_normal = pd.DataFrame(x_scaled , columns = symbols , index = symbols)
df_normal.to_csv('symbols-correlation-normal.csv')
df_normal.head()
```

```python

d_ = pd.read_csv('symbols-correlation-normal.csv', index_col = 'Symbol')
df_stack = d_.where(np.tril(np.ones(df_corr.shape) , k = -1).astype(np.bool))

df_stack = df_stack.stack().reset_index()
df_stack.columns =['Symbol_1', 'Symbol_2','Correlation']

df_stack.to_csv('symbols-correlation-stack.csv')
df_stack.head()
```
