```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('stocks/AAPL.csv', index_col='Date')
df.head()
```

```python
sns.set(rc={'figure.figsize':(14, 4)})
df.loc['2019-01-01':'2019-05-01','Close'].plot(linewidth=0.8);
df.loc['2019-01-01':'2019-05-01','Adj Close'].plot(linewidth=1.5);
plt.legend()
plt.savefig('fig/AAPL-Close-vs-Adj-Closed.svg', format='svg', dpi=1200 , bbox_inches='tight', pad_inches=0)
```

```python
sns.set(rc={'figure.figsize':(14, 4)})
df.loc['2015-01-01':'2020-03-31','Adj Close'].plot(linewidth=0.5);
plt.savefig('fig/APPL-fulltimeseries.svg', format='svg', dpi=1200 ,bbox_inches='tight', pad_inches=0)
```

```python
sns.set(rc={'figure.figsize':(14, 4)})
df.loc['2019-01-01':'2020-03-31','Adj Close'].plot(linewidth=0.8 , label = 'Price');
df.loc['2019-01-01':'2020-03-31','Adj Close'].rolling(5).mean().plot(linewidth=1.5, label = 'Moving average price');
plt.legend()
plt.savefig('fig/APPL-short-period-MA.svg', format='svg', dpi=1200 , bbox_inches='tight', pad_inches=0)
```
