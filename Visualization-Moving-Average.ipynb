```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('stocks/total_data_changePercent_Date.csv', index_col='Date')
df.head()
```

```python
df.shape
```

```python
# check for null value
df.columns[df.isna().any()].tolist()
```

```python
sns.set(rc={'figure.figsize':(14, 4)})
df['A'].plot(linewidth=0.5);
```

```python
sns.set(rc={'figure.figsize':(14, 4)})
df.loc['2015-01-01':'2015-12-30','A'].plot(linewidth=0.8);
df.loc['2015-01-01':'2015-12-30','A'].rolling(10).mean().plot(linewidth=1);
```

```python
sns.set(rc={'figure.figsize':(14, 8)})
df.loc['2019-01-01':'2019-12-30'][['CAT','CMI']].plot(linewidth=0.5 , subplots=True);
```
